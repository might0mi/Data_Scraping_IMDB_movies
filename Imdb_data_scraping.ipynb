```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
urls=["https://www.imdb.com/search/title/?genres=drama&groups=top_250&sort=user_rating,desc&ref_=adv_prv","https://www.imdb.com/search/title/?genres=drama&groups=top_250&sort=user_rating,desc&start=51&ref_=adv_nxt","https://www.imdb.com/search/title/?genres=drama&groups=top_250&sort=user_rating,desc&start=101&ref_=adv_nxt","https://www.imdb.com/search/title/?genres=drama&groups=top_250&sort=user_rating,desc&start=151&ref_=adv_nxt"]

all_pages=[]

for url in urls:
    page=requests.get(url)
    soup=BeautifulSoup(page.text,"html.parser")
    all_pages.append(soup.find_all(attrs={"lister-item mode-advanced"}))
```

```python
index=[]
name=[]
year=[]
certificate=[]
run_time=[]
genre=[]
rating=[]
summary=[]
director=[]
stars=[]
gross=[]
```

```python
for content in all_pages:
    for row in content:

        index.append(row.find_all(class_="lister-item-index unbold text-primary")[0].find(text=True))

        name.append(row.find_all("a")[1].find(text=True))

        year.append(row.find_all(class_="lister-item-year text-muted unbold")[0].find(text=True))

        try:
            certificate.append((row.find_all(attrs={"class":"certificate"})[0]).find(text=True))
        except:
            certificate.append("No Data")

        run_time.append(row.find_all(class_="runtime")[0].find(text=True))

        genre.append(str(row.find_all(class_="genre")[0].find(text=True)))

        rating.append(row.find_all("strong")[0].find(text=True))

        summary.append(str(row.find_all(class_="text-muted")[2].find(text=True)))

        director.append((row.find_all("a")[13].find(text=True)))

        a=(row.find_all("a")[13].find(text=True))
        b=(row.find_all("a")[14].find(text=True))
        c=(row.find_all("a")[15].find(text=True))
        d=(row.find_all("a")[16].find(text=True))

        stars.append(str(a)+","+ str(b) +","+ str(c) +","+ str(d))

        try:
            gross.append((row.find_all(attrs={"name":"nv"})[1]).find(text=True))
        except:
            gross.append("No Data")
```

```python
data=pd.DataFrame()

data["Index"]=index
data["Movie_Name"]=name
data["Release_Year"]=year
data["Certificate"]=certificate
data["Run_Time"]=run_time
data["Rating"]=rating
data["Summary"]=summary
data["Director"]=director
data["Actors"]=stars
data["Gross_Income"]=gross
```

```python
data["Release_Year"]=data["Release_Year"].str[1:-1]
data["Summary"]=data["Summary"].str[3:]
```

```python
data.to_csv("imdb.csv")
```
